```python
import pandas as pd
import io
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

# Dummy data to avoid FileNotFoundError
csv_data = """RowNumber,Surname,Geography,Gender,CreditScore,Exited
1,Hargrave,France,Female,619,1
2,France,Female,608,0
3,Germany,Male,502,1
4,Spain,Female,699,0
5,France,Male,850,0
"""
df = pd.read_csv(io.StringIO(csv_data), index_col="RowNumber")
df.head()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df=df.drop(['Surname', 'Geography','Gender'], axis=1)
scaler=StandardScaler()
df=pd.DataFrame(scaler.fit_transform(df))
df.head()
```

```python
X,Y=df.iloc[:,:-1].values ,df.iloc[:,-1].values
print('Input:\n',X,'\nOutput:\n',Y)
Xtrain,Xtest,Ytrain,Ytest = train_test_split(X, Y, test_size=0.2)
print("Xtrain:\n" ,Xtrain, "\nXtest:\n", Xtest)
print("\nYtrain:\n" ,Ytrain, "\nYtest:\n", Ytest)
```
